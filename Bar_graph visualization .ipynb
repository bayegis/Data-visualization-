```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('C:\\Users\\HP\\Desktop\\INDAKA\\case_time_series.csv')

# Create a bar graph of the total cases over time
sns.barplot(data=df, x='Date', y='Total Confirmed')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.title('Total Confirmed Cases Over Time')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Set the interval between x-axis labels to 9 days
plt.xticks(range(0, len(df['Date']), 9))

# Save the plot as an image file
plt.savefig("G:\\COVID19.jpeg")

# Show the plot
plt.show()
```
