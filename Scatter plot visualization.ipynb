```python
#Scatter plot 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

# Read the CSV file
df = pd.read_csv('C:\\Users\\HP\\Desktop\\INDAKA\\case_time_series.csv')

# Create a scatter plot of the total confirmed cases vs. the daily confirmed cases
sns.scatterplot(data=df, x='Total Confirmed', y='Daily Confirmed')

# Set labels and title
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Daily Confirmed Cases')
plt.title('Scatter Plot of Total Confirmed Cases vs. Daily Confirmed Cases')

# Show the plot
plt.show()
```
