```python
#COVID-19 Data Visualization using matplotlib in Python. 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import os

# Read the CSV file
df = pd.read_csv('C:\\Users\\HP\\Desktop\\INDAKA\\case_time_series.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Create a line plot of the total cases over time
sns.lineplot(data=df, x='Date', y='Daily Confirmed')
plt.xticks(range(0, len(df['Date']), 16))

# Show the plot
plt.show()
```
